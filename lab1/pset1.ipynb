```python
### Problem Set 1
import pandas as pd
import matplotlib.pyplot as plt
import math

# input the dataset into notebook
data = pd.read_csv('ProblemSet1_data.csv')
```

```python
## Descriptive Statistics
# Plot the histograms of the data from experiments A-D (2x2 array using the subplot function)
# expta = data["Expt A"]
# print(expta)
num = 4
bins = 10
for i in range(num):
    expt_name = data.columns[i + 1]
    plt.subplot(2, 2, i + 1, xlabel=expt_name)
    plt.hist(data[expt_name], bins)

plt.tight_layout()
```

```python
# Plot the histograms of the data from experiments A-D (2x2 array using the subplot function) with the bin widths given by the Sturges' rule
num = 4
num_data = data.shape[0]
print(num_data)
bins = int(1 + math.log2(num_data))
for i in range(num):
    expt_name = data.columns[i + 1]
    plt.subplot(2, 2, i + 1, xlabel=expt_name)
    plt.hist(data[expt_name], bins)

plt.tight_layout()
```

Modalities of the experiments
Experiments A, B, and D are unimodal whereas experiment C is bimodal.

```python
# Show the boxplots of all 4 data sets in a single figure
data.boxplot(data.columns[1:].tolist())
```

```python
# Show the scatter plots for all 4 data sets, with the x axis being the experiment number
# print(data.columns)
for i in range(num):
    expt_name = data.columns[i + 1]
    data.plot.scatter(x='Expt #', y=expt_name)
```

```python
## Find the mean, median, range, variance, standard deviation, and IQR for all datasets
no_trials_df = data.drop('Expt #', axis=1)

# Mean
print("MEAN:")
print(no_trials_df.mean())
print()

# Median
print("MEDIAN:")
print(no_trials_df.median())
print()

# Range
print("RANGE")
print(no_trials_df.max() - no_trials_df.min())
print()

# Variance
print("VARIANCE:")
print(no_trials_df.var())
print() 

# Standard Deviation
print("STANDARD DEVIATION:")
print(no_trials_df.std())
print()

# IQR
print("Inter Quartile Range")
print(no_trials_df.quantile(0.75) - no_trials_df.quantile(0.25))
```

```python
## Show the values that are in the top 10 percentile

# get cut off for top 10
top10 = no_trials_df.quantile(0.9)
for i in range(num):
    expt_name = data.columns[i + 1]
    p90 = top10[expt_name]
    greater = data[expt_name] > p90
    print(data[expt_name][greater])
```
